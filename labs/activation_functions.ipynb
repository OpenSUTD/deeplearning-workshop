```
import tensorflow.compat.v2 as tf
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

# Exploring Activation Functions

```
def plot_act(act):
    x = tf.range(-5.0, 5.0, delta=0.01, dtype=tf.float32, name='x_axis')
    with tf.GradientTape() as t:
        t.watch(x)
        y = act(x)
    dy_dx = t.gradient(y, x)

    plt.title(act.__name__+" function and derivative")
    plt.plot(x, y, label="f(x)")
    plt.plot(x, dy_dx, label="f'(x)")
    plt.axhline(y=0, linestyle="dashed", c="k", alpha=0.5)
    plt.legend()
    plt.show()
```

## Built-in Activation Functions

```
plot_act(tf.keras.activations.sigmoid)
```

```
plot_act(tf.keras.activations.tanh)
```

```
plot_act(tf.keras.activations.relu)
```

## Newer Activation Functions

```
def gelu(x):
    """
    Gaussian Error Linear Unit.
    https://arxiv.org/abs/1606.08415
    """
    cdf = 0.5 * (1.0 + tf.math.erf(x / 1.4142135623730951))
    return x * cdf

def swish(x):
    """
    Swish: Self-Gated Activation Function, discovered via reinforcement learning.
    https://arxiv.org/abs/1710.05941
    """
    return x * tf.math.sigmoid(x)
```

```
plot_act(gelu)
```

```
plot_act(swish)
```
